```python
# 2.  Fungsi / Functions
def hitung_efisiensi(input_energi, output_energi):
    efisiensi = (output_energi / input_energi) * 100

    return efisiensi

hasil = hitung_efisiensi(500, 400)
print("efisiensi turbin =", hasil, "%")
```

**Penjelasan :**
Program ini menghasilkan nilai efisiensi turbin, yaitu perbandingan antara output energi dengan input energi.
Ketika dijalankan dengan nilai input energi sebesar 500 dan output energi sebesar 400, program menampilkan hasil efisiensi sebesar 80 %. Artinya, dari total energi masukan sebesar 500, hanya 400 yang berhasil dikonversi menjadi energi keluaran. Dengan demikian, turbin memiliki efisiensi 80%, dan sisanya hilang sebagai panas atau energi lain.

**Cara Kerja Program :**
1. Program mendefinisikan sebuah fungsi bernama hitung_efisiensi() yang digunakan untuk menghitung efisiensi energi pada turbin.
2. Di dalam fungsi, dilakukan perhitungan dengan rumus: Efisiensi = (Output Energi / Input Energi)
3. Nilai hasil perhitungan dikembalikan (return) ke pemanggil fungsi.
4. Program kemudian memanggil fungsi tersebut dengan input energi = 500 dan output energi = 400.
5. Hasil efisiensi yang diperoleh disimpan ke dalam variabel hasil dan ditampilkan di layar dalam satuan persen (%).

**Kode Penting Program :**
1. def hitung_efisiensi(input_energi, output_energi): Baris ini digunakan untuk mendefinisikan fungsi bernama hitung_efisiensi yang memiliki dua parameter, yaitu output energi dan input energi  
2. efisiensi = output_energi / input_energi:  Baris ini adalah inti perhitungan program. Di sini dilakukan pembagian antara output energi dan input energi untuk mendapatkan nilai efisiensi.
3. return efisiensi : Baris ini berfungsi untuk mengembalikan hasil perhitungan dari fungsi ke bagian utama program agar nilainya bisa disimpan atau ditampilkan.
4. hasil = hitung_efisiensi(500, 400) : Baris ini adalah pemanggilan fungsi. Nilai 500 dimasukkan sebagai energi input, dan 400 sebagai energi output. Hasil dari perhitungan efisiensi disimpan ke dalam variabel hasil.
5. print("efisiensi turbin =", hasil, "%") :  Baris ini berfungsi untuk menampilkan hasil efisiensi ke layar agar pengguna dapat melihat persentase efisiensi turbin yang telah dihitung.

```python
# 4. Logika if Majemuk/bertingkat
tekanan_uap = 160
status_manual_override = False

if tekanan_uap > 150 and status_manual_override == False:
    print("Katup Terbuka")
else:
    print("Katup Tertutup")
```

**Penjelasan :**
Program ini digunakan untuk menentukan status katup (valve) berdasarkan nilai tekanan uap dan kondisi manual override. Pada kondisi yang diberikan, yaitu tekanan_uap = 160 dan status_manual_override = False, program menampilkan hasil Katup Terbuka. Artinya, karena tekanan uap melebihi 150 dan sistem tidak sedang dalam mode manual, maka katup otomatis terbuka untuk mengurangi tekanan berlebih.

**Cara Kerja Program :**
1. Program mendefinisikan dua variabel, yaitu tekanan_uap (dengan nilai 160) dan status_manual_override (dengan nilai False).
2. Percabangan if digunakan untuk memeriksa dua kondisi sekaligus:
- Apakah tekanan uap lebih besar dari 150.
- Dan apakah status manual override False (artinya sistem otomatis aktif).
3. Jika kedua kondisi benar, maka program menampilkan “Katup Terbuka”.
4. Jika salah satu kondisi tidak terpenuhi (misalnya tekanan tidak tinggi atau manual override aktif), maka bagian else dijalankan dan program menampilkan “Katup Tertutup”.

**Kode Penting Program :**
1. if tekanan_uap > 150 and status_manual_override == False:  kondisi utama untuk menentukan apakah katup harus terbuka.
2. print("Katup Terbuka")  perintah untuk menampilkan status ketika tekanan tinggi dan sistem otomatis.
3. else:  dijalankan jika kondisi sebelumnya tidak terpenuhi.
4. print("Katup Tertutup")  menampilkan status ketika katup dalam kondisi tertutu

```python
# 10. While Loop & Logika
kapasitas = 100
isi_tangki = 0

while isi_tangki <= kapasitas:
    print("Isi tangki saat ini:", isi_tangki, "liter")

    if isi_tangki == 80:
        print("WARNING: Hampir penuh!")
    elif isi_tangki == 100:
        print("TANGKI PENUH – Pengisian berhenti.")
        break

    isi_tangki += 10
```

**Penjelasan :** Program ini menghasilkan tampilan proses pengisian tangki air secara bertahap dari 0 liter hingga penuh (100 liter). Selama proses berlangsung, program menampilkan volume tangki setiap 10 liter sekali. Saat isi tangki mencapai 80 liter, muncul peringatan “WARNING: Hampir penuh!”, dan ketika mencapai 100 liter, program menampilkan pesan “TANGKI PENUH Pengisian berhenti.” lalu proses berhenti secara otomatis.

**Cara Kerja Program :**
1. Program dimulai dengan menentukan kapasitas tangki sebesar 100 liter dan isi awal 0 liter.
2. Perintah while isi_tangki <= kapasitas: membuat program terus berjalan selama isi tangki belum melebihi kapasitas maksimum.
3. Di dalam perulangan, program menampilkan jumlah isi tangki setiap iterasi.
4. Kondisi if–elif digunakan untuk memeriksa:
- Jika isi tangki mencapai 80 liter, tampilkan peringatan “Hampir penuh”.
- Jika isi tangki mencapai 100 liter, tampilkan pesan “Tangki penuh” dan hentikan program dengan break.
5. Setelah setiap perulangan, nilai isi_tangki bertambah 10 liter (isi_tangki += 10), sehingga simulasi berjalan bertahap sampai penuh.

**Kode Penting Program :**
1. while isi_tangki <= kapasitas:  digunakan untuk menjalankan pengulangan selama tangki belum penuh.
2. if isi_tangki == 80:  memberikan peringatan saat isi tangki mendekati batas maksimum.
3. elif isi_tangki == 100:  mendeteksi kondisi tangki penuh dan menghentikan perulangan dengan break.
4. isi_tangki += 10 : menambahkan isi tangki sebesar 10 liter setiap kali perulangan berjalan, sehingga simulasi pengisian terlihat bertahap.

```python
# Challenge
def hitung_energi(daya, waktu):
    energi = (daya * waktu) / 1000  # hasil dalam kWh
    return energi

print("=== Pemantau Pemakaian Charger HP ===")

daya_charger = float(input("Masukkan daya charger (Watt): "))
waktu_pakai = float(input("Masukkan lama pengecasan (jam): "))

energi = hitung_energi(daya_charger, waktu_pakai)
print("Energi yang digunakan:", energi, "kWh")

# Menentukan status efisiensi
if energi < 0.01:
    print("Status: Sangat hemat ")
elif energi <= 0.05:
    print("Status: Normal ")
else:
    print("Status: Boros! Cabut charger bila tidak dipakai ")

# Simulasi lama pengecasan dengan for loop
for jam in range(1, int(waktu_pakai) + 1):
    print("Jam ke-", jam, ": HP masih dicas...")

# Simulasi pengisian dengan while loop
baterai = 0
while baterai < 100:
    baterai += 25
    print("Baterai terisi:", baterai, "%")
    if baterai == 100:
        print("Pengisian selesai! Cabut charger ")
```

**Penjelasan :** Program ini menghasilkan informasi tentang pemakaian energi listrik pada charger HP.
Program menghitung berapa besar energi (dalam kWh) yang digunakan selama proses pengecasan, lalu memberikan status efisiensi apakah hemat, normal, atau boros.
Selain itu, program juga mensimulasikan proses pengecasan HP dari awal hingga baterai penuh menggunakan perulangan for dan while.

**Cara Kerja Program :**  
1. Program dimulai dengan mendefinisikan fungsi hitung_energi() yang menghitung energi listrik dengan rumus (daya × waktu) / 1000, hasilnya dalam satuan kWh.
2. Pengguna memasukkan daya charger (Watt) dan lama waktu pengecasan (jam).
3. Fungsi dipanggil untuk menghitung energi berdasarkan input tersebut, lalu hasilnya ditampilkan.
4. Dengan menggunakan percabangan if–elif–else, program menentukan status: Energi < 0.01 = Sangat hemat, Energi antara 0.01–0.05 = Normal, Energi > 0.05  = Boros
3. Perulangan for digunakan untuk menampilkan simulasi proses pengecasan tiap jam.
4. Perulangan while digunakan untuk mensimulasikan pengisian baterai dari 0% sampai 100%. Ketika baterai mencapai 100%, program menampilkan pesan “Pengisian selesai! Cabut charger”.

**Kode Penting Program :**
1. def hitung_energi(daya, waktu):  mendefinisikan fungsi untuk menghitung energi listrik.
2. energi = (daya * waktu) / 1000  rumus perhitungan energi dalam kWh.
3. if energi < 0.01:  awal dari percabangan untuk menentukan status efisiensi.
4. for jam in range(1, int(waktu_pakai) + 1):  perulangan yang menampilkan waktu pengecasan.
5. while baterai < 100:  simulasi pengisian baterai hingga penuh.
