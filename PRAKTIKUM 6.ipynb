```python
# SOAL 2: Menghitung Volume Balok
def hitung_volume_balok(panjang, lebar, tinggi):
    """Menghitung volume balok dari panjang, lebar, dan tinggi."""
    return panjang * lebar * tinggi


# Pemanggilan fungsi
panjang = 4
lebar = 3
tinggi = 4

print(f"Volume balok dengan ukuran {panjang} x {lebar} x {tinggi} adalah: {hitung_volume_balok(panjang, lebar, tinggi)}")
```

```python
# SOAL 3: Menghitung Rata-rata dari List Nilai
def hitung_rata_rata(data_nilai):
    "Menghitung rata-rata dari list angka menggunakan perulangan for"
    total = 0
    jumlah_data = 0

    # Perulangan untuk menjumlahkan nilai valid
    for nilai in data_nilai:
        # Pengecekan nilai valid (0 - 100)
        if nilai < 0 or nilai > 100:
            # Nilai tidak valid, diabaikan
            continue
        else:
            total += nilai
            jumlah_data += 1

    # Pembagian untuk rata-rata
    if jumlah_data > 0:
        return total / jumlah_data
    else:
        return 0

# Pemanggilan fungsi
nilai_praktikan1 = [85, 78, 92, 65, 90]
print(f"Nilai rata-rata kelas adalah: {hitung_rata_rata(nilai_praktikan1)}")
```

```python
# SOAL 4: Program Interaktif Penentuan Predikat Nilai

def tentukan_predikat(skor):
    "Menentukan predikat dan keterangan lulus/tidak lulus berdasarkan skor."

    if skor >= 85 and skor <= 100:
        return "Predikat: A | Lulus (Sangat Baik)"
    elif skor >= 70 and skor < 85:
        return "Predikat: B | Lulus (Baik)"
    elif skor >= 60 and skor < 70:
        return "Predikat: C | Lulus (Cukup)"
    elif skor >= 50 and skor < 60:
        return "Predikat: D | Tidak Lulus (Kurang)"
    elif skor >= 0 and skor < 50:
        return "Predikat: E | Tidak Lulus (Sangat Kurang)"
    else:
        return "Skor harus berada antara 0–100."


# Program utama
print("=== SISTEM PENENTUAN PREDIKAT NILAI ===")
skor = float(input("Masukkan skor (0–100): "))
hasil = tentukan_predikat(skor)
print(hasil)
```

```python
# SOAL 4: Sistem Penentuan Predikat Nilai
def tentukan_predikat(skor):
    "Menentukan predikat (A, B, C, D, E) berdasarkan skor."

    # Pengecekan tipe data numerik
    if type(skor) == int or type(skor) == float:
        if skor >= 85:
            return "A"
        elif skor >= 70:
            return "B"
        elif skor >= 60:
            return "C"
        elif skor >= 50:
            return "D"
        else:
            return "E"
    else:
        return "Input tidak valid, masukkan angka."


# Pemanggilan fungsi
print(f"Skor 77 mendapat predikat: {tentukan_predikat(77)}")
print(f"Skor 54 mendapat predikat: {tentukan_predikat(54)}")
print(f"Skor 'lima puluh' mendapat predikat: {tentukan_predikat('lima puluh')}")
```

```python
# SOAL 5: Menghitung Faktorial dengan Loop While
def hitung_faktorial(n):
    """Menghitung faktorial dari bilangan bulat positif n."""

    # Percabangan untuk menangani kasus khusus
    if n < 0:
        return "Faktorial tidak terdefinisi untuk bilangan negatif."
    elif n == 0:
        return 1
    else:
        hasil = 1
        # Perulangan while selama n lebih besar dari 1
        while n > 1:
            hasil *= n
            n -= 1  # Pengurangan nilai n
        return hasil


# Pemanggilan fungsi
bilangan_n = 4
print(f"Faktorial dari {bilangan_n} adalah: {hitung_faktorial(bilangan_n)}")

bilangan_n = 0
print(f"Faktorial dari {bilangan_n} adalah: {hitung_faktorial(bilangan_n)}")

bilangan_n = -9
print(f"Faktorial dari {bilangan_n} adalah: {hitung_faktorial(bilangan_n)}")
```

```python
# SOAL 6: Analisis Data Suhu Ekstrem
def analisis_suhu(data_suhu, batas_panas=35, batas_dingin=15):
    """Menghitung jumlah hari Panas dan Dingin dari data_suhu."""

    panas_count = 0
    dingin_count = 0

    for suhu in data_suhu:
        if suhu >= batas_panas:
            panas_count += 1
        elif suhu <= batas_dingin:
            dingin_count += 1

    return panas_count, dingin_count


# Pemanggilan fungsi
suhu_mingguan = [36, 30, 14, 25, 38, 16, 28]
jumlah_panas, jumlah_dingin = analisis_suhu(suhu_mingguan)

print(f"Jumlah hari Panas (≥35°C): {jumlah_panas}")
print(f"Jumlah hari Dingin (≤15°C): {jumlah_dingin}")
```
