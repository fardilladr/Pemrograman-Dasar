Kode **df.mean(numeric_only=True)** digunakan untuk menghitung nilai rata-rata (mean) dari seluruh kolom numerik pada DataFrame. Parameter **numeric_only=True** memastikan bahwa hanya kolom bertipe angka yang dihitung, sehingga kolom bertipe teks tidak ikut diproses.

```python
# Mode
df.mode()
```

Fungsi **df.mode()** membantu mengidentifikasi nilai yang paling sering muncul dalam setiap kolom. Hasil menunjukkan bahwa kolom kategori dapat memiliki lebih dari satu mode, sehingga Pandas menampilkan beberapa baris untuk mencerminkan seluruh nilai yang memiliki frekuensi tertinggi.


```python
import matplotlib.pyplot as plt

df['price'].hist()
plt.title("Distribusi Harga Produk")
plt.xlabel("Harga")
plt.ylabel("Frekuensi")
plt.show()

df['quantity'].hist()
plt.title("Distribusi Jumlah Stok")
plt.xlabel("Quantity")
plt.ylabel("Frekuensi")
plt.show()
```

Kode tersebut digunakan untuk membuat histogram yang menggambarkan distribusi data pada kolom price dan quantity dalam DataFrame.

**df['price'].hist()**
Membuat histogram untuk menampilkan penyebaran nilai harga produk. Grafik ini menunjukkan seberapa sering setiap rentang harga muncul di dataset.

**df['quantity'].hist()**
Membuat histogram untuk menggambarkan distribusi jumlah stok produk, sehingga terlihat berapa kali setiap rentang jumlah stok muncul.

**plt.title(), plt.xlabel(), dan plt.ylabel**()
Berfungsi memberi judul grafik serta menambahkan label pada sumbu X dan Y agar visualisasi mudah dipahami.

**plt.show()**
Menampilkan grafik histogram ke layar.

```python
# Korelasi
df[['price','quantity']].corr()
```

Kode `df[['price','quantity']].corr()` digunakan untuk menghitung hubungan linear antara kolom harga dan jumlah stok. Hasil korelasi sebesar **0,062** menunjukkan bahwa hubungan antara kedua variabel tersebut **sangat lemah**, sehingga perubahan harga tidak berpengaruh secara signifikan terhadap jumlah stok, dan keduanya dapat dianggap tidak saling berkaitan dalam konteks dataset ini.


```python
# Kovarian
df[['price','quantity']].cov()
```

Nilai kovarian menunjukkan bahwa hubungan antara harga (price) dan jumlah stok (quantity) bernilai 7.086, yang merupakan nilai positif dan relatif kecil. Ini berarti harga dan jumlah stok cenderung bergerak searah, tetapi hubungannya sangat lemah. Sementara itu, nilai kovarian pada diagonal (94.39 untuk harga dan 136.49 untuk stok) menunjukkan variasi internal masing-masing variabel. Secara keseluruhan, kovarian ini menegaskan bahwa meskipun harga dan jumlah stok memiliki arah hubungan yang sama, keterkaitannya hampir tidak signifikan.

```python
df.groupby('product_type')['price'].mean()
```



Kode **df.groupby('product_type')['price'].mean()** digunakan untuk menghitung **rata-rata harga** berdasarkan setiap kategori produk. Hasilnya menunjukkan bahwa **Shirt** memiliki rata-rata harga tertinggi sebesar **108,36**, diikuti oleh **Jacket** dengan rata-rata **107,38**, dan **Trousers** memiliki rata-rata harga terendah yaitu **101,67**. Informasi ini membantu memahami perbedaan harga rata-rata antar jenis produk dalam dataset.


```python
df.groupby('product_type').agg({
    'price': ['mean','max','min'],
    'quantity': ['mean','sum']
})
```

Kode df.groupby('product_type').agg({...}) digunakan untuk menghitung beberapa statistik sekaligus berdasarkan jenis produk. Hasilnya menunjukkan bahwa untuk kolom price, rata-rata harga tertinggi terdapat pada Shirt (108,36), disusul Jacket (107,38), sedangkan Trousers memiliki rata-rata terendah (101,67). Nilai maksimum harga pada semua kategori sama, yaitu 119, sementara nilai minimumnya berada di kisaran 90â€“92. Untuk kolom quantity, rata-rata jumlah stok antar kategori hampir sama, yaitu sekitar 60 unit, dan total stok tertinggi dimiliki Jacket (25.527), diikuti Shirt (25.469), dan Trousers (25.139). Ringkasan ini memberikan gambaran lengkap mengenai perbandingan harga dan jumlah stok berdasarkan tipe produk.

```python
print("Harga minimum :", df['price'].min())
print("Harga maksimum :", df['price'].max())
print("Harga rata-rata :", df['price'].mean())
print("Quantity rata-rata :", df['quantity'].mean())
print("Quantity minimum :", df['quantity'].min())
print("Quantity maksimum :", df['quantity'].max())
```

Kode tersebut digunakan untuk menampilkan nilai minimum, maksimum, dan rata-rata dari kolom price serta nilai minimum, maksimum, dan rata-rata dari kolom quantity. Hasilnya menunjukkan bahwa harga produk berada pada rentang 90 hingga 119, dengan rata-rata sekitar 105,81. Sementara itu, jumlah stok berada pada rentang 40 hingga 80 unit, dengan rata-rata sekitar 60,14 unit. Informasi ini memberikan gambaran umum mengenai kisaran harga dan stok produk dalam dataset.

```python
plt.hist(df['price'])
plt.title("Distribusi 'price'")
plt.show()

plt.hist(df['quantity'])
plt.title("Distribusi 'quantity'")
plt.show()
```

Kode tersebut digunakan untuk membuat histogram yang menampilkan distribusi data pada kolom price dan quantity dalam DataFrame.

**plt.hist(df['price'])**
Membuat histogram untuk melihat penyebaran nilai harga produk serta frekuensi kemunculannya dalam dataset.

**plt.title("Distribusi 'price'")**
Memberikan judul grafik agar visualisasi lebih jelas.

**plt.show()**
Menampilkan grafik histogram harga ke layar.

**plt.hist(df['quantity'])**
Membuat histogram untuk melihat distribusi jumlah stok produk dan frekuensi setiap rentang quantity.

**plt.title("Distribusi 'quantity'")**
Menambahkan judul untuk grafik distribusi quantity.

**plt.show()**
Menampilkan grafik histogram jumlah stok ke layar.

```python
corr = df[['price','quantity']].corr()
cov = df[['price','quantity']].cov()

print("Korelasi:\n", corr)
print("\nKovarian:\n", cov)
```

**Korelasi** mengukur seberapa kuat hubungan linear antara dua variabel, dalam hal ini price dan quantity. Nilai korelasi yang didapat adalah 0.062, yang berarti hubungan keduanya sangat lemah dan hampir tidak ada hubungan linear. Artinya, perubahan pada harga tidak secara signifikan memengaruhi jumlah barang yang dibeli. Korelasi juga distandarkan dengan rentang -1 sampai 1, sehingga mudah dibaca untuk melihat apakah hubungan itu kuat atau lemah.

**Kovarian** menunjukkan arah hubungan antara dua variabel, tanpa mengukur seberapa kuat hubungan tersebut. Pada output, kovarian antara price dan quantity adalah 7.08, yang bernilai positif, sehingga mengindikasikan bahwa ketika harga meningkat, jumlah yang dibeli cenderung ikut meningkat, tetapi hanya dalam jumlah yang sangat kecil. Kovarian tidak distandarkan seperti korelasi, sehingga angkanya lebih sulit dibandingkan langsung, tetapi tetap memberi gambaran arah hubungan kedua variabel.
